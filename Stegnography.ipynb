```python
import cv2
import numpy as np

def get_bin_str(s):
    return ''.join(format(ord(c), '07b') for c in s)

def encode(I, s):
    s += chr(0)
    x = get_bin_str(s)
    x = np.array(list(x), dtype=np.uint8)
    IRound = I - I % 2
    IRound = IRound.flatten()
    IRound[:x.size] += x
    return IRound.reshape(I.shape)

def decode(IEnc):
    places = np.array([2**(6 - i) for i in range(7)])
    IDec = IEnc.flatten()
    i = 0
    s = ""
    while True:
        bits = IDec[i:i+7] % 2
        val = np.sum(places * bits)
        if val == 0:
            break
        s += chr(val)
        i += 7
    return s

# Load the image
I = cv2.imread('rose.jpg', cv2.IMREAD_GRAYSCALE)

# Encode and decode
I2 = encode(I, "Hello,  steganography")
print("Decoded message:", decode(I2))
```
